### [ **X** ] Fit a ridge regression model with multiple features. Use the `normalize=True` parameter (or do [feature scaling](https://scikit-learn.org/stable/modules/preprocessing.html) beforehand — use the scaler's `fit_transform` method with the train set, and the scaler's `transform` method with the test set)

```python
#Time for univariate feature selection
from sklearn.feature_selection import f_regression, SelectKBest

#Initiate model
selector = SelectKBest(score_func=f_regression, k=4)

#Use selector
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
X_train_selected.shape, X_test.shape 

selected_names = selector.get_support()
all_names = X_train.columns
selected_names = all_names[selected_names]
print(selected_names)
```

```python
#Ridge Regression!
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error

for alpha in [0.01, 0.1, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]:
  #Fit model
  print(f'Ridge regression, with alpha={alpha}')
  model = Ridge(alpha=alpha)
  model.fit(X_train_selected, y_train)
  y_pred = model.predict(X_test_selected)

  #Get test mae
  mae = mean_absolute_error(y_test, y_pred)
  print(f'Test mae: ${mae:,.0f}')

#Couldn't get the matplotlib graph to work which would have been nice. I cant wait to automate this with some sort of CV
```
